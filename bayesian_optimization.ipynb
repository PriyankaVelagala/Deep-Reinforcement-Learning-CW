```python
import numpy as np 
import statistics
import time 
from grid_world import grid_world

from bayes_opt import BayesianOptimization
```

# Bayesian Optimization
https://github.com/fmfn/BayesianOptimization/blob/master/examples/basic-tour.ipynb

```python
# EPS Greedy
def eps_greedy_function(alpha, gamma, eps_start, eps_end):
    max_steps = 100
    num_eps = 600
    policy = 'eps_greedy'
    myGridWorld = grid_world(False)
    rewards_per_episode, steps_per_episode, found_flag, Q = myGridWorld.run_q_learning(num_eps,
                                                                                   max_steps,
                                                                                   alpha,
                                                                                   gamma, policy, 
                                                                                   {'eps_start' : eps_start ,
                                                                                    'eps_end' : eps_end})
    
    
    return statistics.mean(rewards_per_episode)
    

eps_greedy_pbounds = {'alpha': (0, 1),
              'gamma': (0, 1),
              'eps_start': (0.6, 1),
              'eps_end': (0, 0.5)
              }

eps_greedy_optimizer = BayesianOptimization(
    f=eps_greedy_function,
    pbounds=eps_greedy_pbounds,
    verbose=1, # verbose = 1 prints only when a maximum is observed, verbose = 0 is silent
    random_state=1,
)



# Softmax
def softmax_function(alpha, gamma, T):
    max_steps = 100
    num_eps = 600
    policy = 'softmax'
    myGridWorld = grid_world(False)
    rewards_per_episode, steps_per_episode, found_flag, Q = myGridWorld.run_q_learning(num_eps,
                                                                                   max_steps,
                                                                                   alpha,
                                                                                   gamma,
                                                                                   policy, 
                                                                                   {'T' : T})
    
    
    return statistics.mean(rewards_per_episode)
    

softmax_pbounds = {'alpha': (0, 1),
              'gamma': (0, 1),
              'T': (100, 150)
              }



softmax_optimizer = BayesianOptimization(
    f=softmax_function,
    pbounds=softmax_pbounds,
    verbose=2, # verbose = 1 prints only when a maximum is observed, verbose = 0 is silent
    random_state=1,
)



```

```python
eps_greedy_optimizer.maximize(n_iter=100, init_points=10)
print(eps_greedy_optimizer.max)
```

```python
softmax_optimizer.maximize(n_iter=100, init_points=10)
```

```python
print(softmax_optimizer.max)
```
