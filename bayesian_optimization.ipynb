```python
import numpy as np 
import statistics
import time 
from grid_world import grid_world

from bayes_opt import BayesianOptimization
```

```python
myGridWorld = grid_world(False)
```

# Bayesian Optimization
https://github.com/fmfn/BayesianOptimization/blob/master/examples/basic-tour.ipynb

```python
def black_box_function(alpha, gamma, eps_start, eps_end):
    max_steps = 100
    num_eps = 600
    policy = 'eps_greedy'
    myGridWorld = grid_world(False)
    rewards_per_episode, steps_per_episode, found_flag, Q = myGridWorld.run_q_learning(num_eps,
                                                                                   max_steps,
                                                                                   alpha,
                                                                                   gamma, policy, 
                                                                                   {'eps_start' : eps_start ,
                                                                                    'eps_end' : eps_end})
    
    
    return statistics.mean(rewards_per_episode)
    

pbounds = {'alpha': (0, 1),
              'gamma': (0, 1),
              'eps_start': (0.6, 1),
              'eps_end': (0, 0.5)
              }



optimizer = BayesianOptimization(
    f=black_box_function,
    pbounds=pbounds,
    verbose=1, # verbose = 1 prints only when a maximum is observed, verbose = 0 is silent
    random_state=1,
)

```

```python
optimizer.maximize(n_iter=100, init_points=10)
```

```python
print(optimizer.max)
# {'target': 1501.6333333333334, 'params': {'alpha': 0.4454654350000322, 'eps_end': 0.11419465170061395, 'eps_start': 0.6, 'gamma': 0.11581353253177103}}
```
